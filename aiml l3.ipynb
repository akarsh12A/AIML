```python
year=range(2000,2008)
apples=[0.895,0.91,0.919,0.926,0.929,0.934,0.936,0.937]
oranges=[0.962,0.941,0.930,0.918,0.908,0.907,0.904,0.901]
plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel('year')
plt.ylabel('yield')
plt.title('Crop yields in Kanto')
plt.legend(['apples','oranges'])
```

```python
plt.plot(year,apples,marker='o')
plt.plot(year,oranges,marker='o')

plt.ylabel('yeild')
plt.title('crop yeilds in kanto')
plt.legend(['apples','oranges']);
```

```python
plt.plot(year,apples,marker='s',c='b',ls='-',lw=2,ms=8,mew=2,mec='navy')
plt.plot(year,oranges,marker='o',c='r',ls='--',lw=3,ms=10,alpha=0.5)
plt.xlabel('years')
plt.ylabel('yeild')
plt.title('crop yeilds in kanto')
plt.legend(['apples','oranges']);
```

```python
plt.plot(year,apples,'s--b',label='apples')
plt.plot(year,oranges,'o--r',label='oranges')
plt.xlabel('years')
plt.ylabel('yeild')
plt.title('crop yeilds in kanto')
plt.legend(['apples','oranges']);
```

```python
plt.plot(year,apples,'sb',label='apples')
plt.plot(year,oranges,'or',label='oranges')
plt.xlabel('years')
plt.ylabel('yeild')
plt.title('crop yeilds in kanto')
plt.legend(['apples','oranges']);
```
